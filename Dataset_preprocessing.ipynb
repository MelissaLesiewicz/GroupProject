```python
# Initial imports
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
filepath="Resources/healthcare-dataset-stroke-data.csv"
df = pd.read_csv(filepath)
df.head()
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
#Drop rows with null values (201 rows)
df.dropna(inplace=True)
#Dropping id column
df.drop(columns="id", axis=1, inplace=True)
df.head()
```

```python
y=df["stroke"]
X=pd.get_dummies(df.drop(columns="stroke", axis=1))
X.head()
```

```python
X.describe()
```

```python
y.value_counts()
```

```python
# Standardize the data with MinMaxScaler().
scaler=MinMaxScaler()
scaler.fit(X[["age", "avg_glucose_level", "bmi"]])
X_scaled=scaler.transform(X[["age", "avg_glucose_level", "bmi"]])
X_scaled
```

```python
X_scaled = pd.DataFrame(data=X_scaled, columns=["age", "avg_glucose_level", "bmi"], index=X.index)
X_scaled = X_scaled.join(X.drop(columns=["age", "avg_glucose_level", "bmi"], axis=1))
X_scaled.head()
```

```python
# Use sklearn to split dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, stratify=y)
```

```python
#Exporting testing dataset 
output_file_path = "Resources/stroke_testing_data.csv"
#Creating the dataframe 
testing_df = pd.DataFrame(data=X_test, columns=X_scaled.columns)
testing_df["stroke"]=y_test
#Creating csv file 
testing_df.to_csv(output_file_path, index=False)
```

# Sampling Techniques

```python
from collections import Counter
# Resample the training data with the RandomOversampler
from imblearn.over_sampling import RandomOverSampler

ros=RandomOverSampler()
X_randomoversample, y_randomoversample = ros.fit_resample(X_train, y_train)
Counter(y_randomoversample)
```

```python
# Resample the data using the ClusterCentroids resampler
from imblearn.under_sampling import ClusterCentroids

cc = ClusterCentroids()
X_clustercentroid, y_clustercentroid = cc.fit_resample(X_train, y_train)
Counter(y_clustercentroid)
```

```python
# Resample the training data with SMOTEENN
from imblearn.combine import SMOTEENN

smote_enn = SMOTEENN()
X_smoteen, y_smoteen = smote_enn.fit_resample(X_train, y_train)
Counter(y_smoteen)
```

# Creating training and testing data sets

```python
#Exporting ROS dataset
output_file_path = "Resources/stroke_training_randomoversampled.csv"
#Creating the dataframe 
ros_training_df = pd.DataFrame(data=X_randomoversample, columns=X_scaled.columns)
ros_training_df["stroke"]=y_randomoversample
#Creating csv file 
ros_training_df.to_csv(output_file_path, index=False)
```

```python
#Exporting SMOTEENN dataset
output_file_path = "Resources/stroke_training_SMOTEENN.csv"
#Creating the dataframe 
smoteenn_training_df = pd.DataFrame(data=X_smoteen, columns=X_scaled.columns)
smoteenn_training_df["stroke"]=y_smoteen
#Creating csv file 
smoteenn_training_df.to_csv(output_file_path, index=False)
```
