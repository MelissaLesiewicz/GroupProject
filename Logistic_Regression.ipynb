```python
def train_test_logistic(X_train, y_train, X_test, y_test, label):

    ## Creating the decision tree classifier instance
    global logic_re
    logic_re = LogisticRegression(penalty='l2', solver='sag', max_iter=100)
    # Fitting the model
    logic_re = logic_re.fit(X_train, y_train)
    # Making predictions using the testing data
    predictions = logic_re.predict(X_test)
    
    # Calculating the confusion matrix
    cm = confusion_matrix(y_test, predictions)
    cm_df = pd.DataFrame(
        cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
    )

    # Calculating the accuracy score
    acc_score = accuracy_score(y_test, predictions)
    
    # Displaying results
    print("Logistic Regression")
    print(f"Confusion Matrix:{label}")
    display(cm_df)
    print(f"Accuracy Score : {acc_score}")
    print("Classification Report")
    print(classification_report(y_test, predictions))
    

```

```python
for i, _ in enumerate(ys):
    X_train = Xs[i]
    y_train= ys[i]
    label= cmlabel[i]
    train_test_logistic(X_train, y_train, X_test, y_test, label)
```

```python
X_train = Xs[0]
y_train= ys[0]
label= cmlabel[0]
train_test_logistic(X_train, y_train, X_test, y_test, label)
```
