```python
# Initial imports.
import pandas as pd
from path import Path
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Data Import
stroke_file_test = Path("Resources/stroke_testing_data.csv")
stroke_file_raw = Path("Resources/stroke_training_ActualDataSample.csv")
stroke_file_oversample = Path("Resources/stroke_training_randomoversampled.csv")
stroke_file_smote = Path("Resources/stroke_training_SMOTEENN.csv")

testing_df = pd.read_csv(stroke_file_test)
training_raw_df = pd.read_csv(stroke_file_raw)
training_oversample_df = pd.read_csv(stroke_file_oversample)
training_smote_df = pd.read_csv(stroke_file_smote)

#testing_df.head()
training_raw_df.head()
#training_oversample_df.head()
#training_smote_df.head()
```

```python
# Define the data for raw training.
raw_X = training_raw_df.copy()
raw_X = raw_X.drop("stroke", axis=1)
raw_X.head()

raw_y = training_raw_df["stroke"]
```

```python
# Define the data for oversampled training.
os_X = training_oversample_df.copy()
os_X = os_X.drop("stroke", axis=1)
os_X.head()

os_y = training_oversample_df["stroke"]
```

```python
# Define the data for SMOTE training.
smote_X = training_smote_df.copy()
smote_X = smote_X.drop("stroke", axis=1)
smote_X.head()

smote_y = training_smote_df["stroke"]
```

```python
X_test = testing_df.copy()
X_test = X_test.drop("stroke", axis=1)

y_test = testing_df["stroke"]
```

#### Model Three: SMOTEEN

```python
# Create model instance
rf_model = RandomForestClassifier(n_estimators=512) 
```

```python
# Fit the model
rf_model = rf_model.fit(smote_X, smote_y)
```

```python
# Making predictions
predictions = rf_model.predict(X_test)
```

```python
# confusion matrix.
cm = confusion_matrix(y_test, predictions)

cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
```

```python
# accuracy score.
acc_score = accuracy_score(y_test, predictions)
```

```python
# results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```
